# Handling outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Assuming df is your DataFrame with numerical columns
# Get all numerical columns
numerical_columns = train_data.select_dtypes(include=['float64', 'int64']).columns

# Plot box plot for each numerical column
plt.figure(figsize=(10, len(numerical_columns) * 5))  # Adjusting the figure size

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 1, i)  # Create subplots for each column
    sns.boxplot(data=train_data, y=col)  # Set y=col to display vertically
    plt.title(f'Box plot of {col}', fontsize=14)
    plt.ylabel(col, fontsize=12)

plt.tight_layout()  # Ensure plots do not overlap
plt.show()
```

```python
#I can see outliers for col0,col2, col3, col4,col5, col6, col7, col8,col9,col10,col14,col15,col16,col17,col18,col19,col20,col21
```

```python
#Handling those outliers using cap or clip outliers method 
Q1 = train_data['Column0'].quantile(0.25)  # First quartile (25th percentile)
Q3 = train_data['Column0'].quantile(0.75)  # Third quartile (75th percentile)
IQR = Q3 - Q1  # Interquartile Range (IQR)

# Calculate the lower bound and upper bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
train_data['Column0'] = train_data['Column0'].apply(lambda x: lower_bound if x < lower_bound else x)

# Cap upper values
train_data['Column0'] = train_data['Column0'].apply(lambda x: upper_bound if x > upper_bound else x)
```

```python
list = ['Column0','Column2','Column3','Column4','Column5','Column6','Column7','Column8','Column9','Column10','Column14','Column15','Column16','Column17','Column18','Column19','Column20','Column21']
for i in list:
    Q1 = train_data[i].quantile(0.25)  # First quartile (25th percentile)
    Q3 = train_data[i].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile Range (IQR)

    # Calculate the lower bound and upper bound
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    train_data[i] = train_data[i].apply(lambda x: lower_bound if x < lower_bound else x)

    # Cap upper values
    train_data[i] = train_data[i].apply(lambda x: upper_bound if x > upper_bound else x)
```

```python
# Assuming df is your DataFrame with numerical columns
# Get all numerical columns
numerical_columns = train_data.select_dtypes(include=['float64', 'int64']).columns

# Plot box plot for each numerical column
plt.figure(figsize=(10, len(numerical_columns) * 5))  # Adjusting the figure size

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 1, i)  # Create subplots for each column
    sns.boxplot(data=train_data, y=col)  # Set y=col to display vertically
    plt.title(f'Box plot of {col}', fontsize=14)
    plt.ylabel(col, fontsize=12)

plt.tight_layout()  # Ensure plots do not overlap
plt.show()
```

```python
import imblearn
from imblearn.over_sampling import SMOTE
```

```python
x = train_data[['Column0', 'Column1', 'Column2', 'Column3', 'Column4', 'Column5', 
                'Column6', 'Column7', 'Column8', 'Column9', 'Column10', 'Column11',
                'Column12', 'Column13', 'Column14', 'Column15', 'Column16', 'Column17', 
                'Column18', 'Column19', 'Column20', 'Column21']]
y = train_data[['target']]
```

```python
smote = SMOTE(random_state=42)
```

```python
x_train_resampled, y_train_resampled = smote.fit_resample(x, y)
```

```python
print(x_train_resampled)
```

```python
print(x_train_resampled.dtypes)
```

```python
print("\nAfter SMOTE:")
print(y_train_resampled.value_counts())
```

```python
x_test = x_test[['Column0', 'Column1', 'Column2', 'Column3', 'Column4', 'Column5', 
                'Column6', 'Column7', 'Column8', 'Column9', 'Column10', 'Column11',
                'Column12', 'Column13', 'Column14', 'Column15', 'Column16', 'Column17', 
                'Column18', 'Column19', 'Column20', 'Column21']]
y_test = y_test[['target']]
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.regularizers import l2
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Assuming x_train_resampled, y_train_resampled, x_test, and y_test are already defined

# Step 1: Scale the data
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_resampled)
x_test_scaled = scaler.transform(x_test)

# Step 2: Create a more complex ANN model
model = keras.Sequential([
    layers.Dense(128, input_shape=(x_train_scaled.shape[1],), kernel_regularizer=l2(0.001)),
    layers.LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    
    layers.Dense(64, kernel_regularizer=l2(0.001)),
    layers.LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    
    layers.Dense(32, kernel_regularizer=l2(0.001)),
    layers.LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),
    layers.Dropout(0.5),

    layers.Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Step 3: Compile the model with RMSprop optimizer
model.compile(optimizer=keras.optimizers.RMSprop(learning_rate=0.001), 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

# Step 4: Define callbacks for early stopping and learning rate reduction
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)

# Step 5: Fit the model with early stopping and learning rate scheduler
model.fit(x_train_scaled, y_train_resampled, epochs=50, batch_size=16, validation_split=0.2, 
          callbacks=[early_stopping, lr_scheduler])

# Step 6: Make predictions
y_pred_prob = model.predict(x_test_scaled)
y_pred = (y_pred_prob > 0.5).astype(int)  # Convert probabilities to binary predictions

# Step 7: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print(classification_report(y_test, y_pred))

```
